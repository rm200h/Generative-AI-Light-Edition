# Instructions for setting up kendra for RAG DEMO

### 1) Download the file kendra-docs-index.yaml, that is present in the same folder as this notebook, into your local machine

### 2) In the AWS console page, search for CloudFormation

### 3) Click cloud formation service

* Click on the Create stack button in the upper right corner of the cloud formation page
* Choose the option With new resources (Standard)


### 4) Under specify template choose upload a template file. Upload the YAML file downloaded on local machine, and click on next. 

### 5) Click Next to name the stack as kendra-rag-stack, click Next. Accept default values and click next again.

### 6) Before sumbitting the cloudformation, it will ask you to acknowledge that you understand stack will create IAM components. Accept it and Click Submit. 

### 7) The stack creation is going to take 15-20 mins 12:16 pm
